## 🧹 Tratamiento de Valores Nulos
Para garantizar que el análisis sea estadísticamente significativo y técnicamente eficiente, se han aplicado las siguientes reglas de procesamiento sobre el dataset:

Identificar las columnas que presentan más del 30 % de valores nulos para decidir su eliminación.
Adicional, se reviso las columnas por la varianza para eliminar aquellas que no sean necesarias.
En el caso de columnas críticas con pocos valores faltantes, se realiza la imputación de datos.

El código **analiza los valores nulos**, elimina las columnas con **más del 30 % de datos faltantes** y muestra un **comparativo del número de columnas antes y después** del proceso de limpieza.


```python
# Obtiene el número total de columnas del DataFrame original
total_columnas_antes = df.shape[1]
total_columnas_antes  # Muestra el total de columnas antes del tratamiento

# Calcula el porcentaje de valores nulos por cada columna
nulos = (
    df.isna()          # Identifica valores nulos (NaN)
      .mean()          # Calcula la proporción de nulos por columna
      .mul(100)        # Convierte la proporción a porcentaje
      .round(2)        # Redondea a 2 decimales
      .reset_index()   # Convierte el índice en una columna
)

# Renombra las columnas del DataFrame resultante
nulos.columns = ['Columna', 'Porcentaje_nulos']

# Selecciona las columnas que tienen más del 30% de valores nulos
columnas_mayor_30 = (
    nulos[nulos['Porcentaje_nulos'] > 30]['Columna']
    .tolist()          # Convierte los nombres de columnas a una lista
)

# Elimina del DataFrame las columnas con más del 30% de nulos
df_reducido = df.drop(columns=columnas_mayor_30)

# "Las columnas que tienen un valor único... probablemente sean inútiles"
for col in df_reducido.columns:
    if df_reducido[col].nunique() == 1:
        print(f"Eliminando columna de varianza cero: {col}")
        df_reducido = df_reducido.drop(columns=[col])

print(f"Dimensiones post-limpieza básica: {df_reducido.shape}")

# Obtiene el total de columnas después de eliminar las columnas con muchos nulos
total_columnas_despues = df_reducido.shape[1]
total_columnas_despues  # Muestra el total de columnas después del tratamiento

# Crea un DataFrame resumen con el antes y después del tratamiento de columnas
resumen_columnas = pd.DataFrame({
    'Estado': ['Antes del tratamiento', 'Después del tratamiento'],
    'Total_columnas': [total_columnas_antes, total_columnas_despues]
})

# Muestra el resumen final
resumen_columnas
```

Luego de validar, nos percatamos que no existen columnas para eliminar.

##  🧹 Tratamiento de Outliers

Al analizar el dataset nos percatamos que en las columnas seleccionadas tenmos valores atípicos que nos darian un problema en el siguiente paso que es el entrenamiento mediante KMeans. Por ello, vamos a realizar la limpieza mediante el método de Rango Intercuartílico (IQR).

```python
# Variables de interés
columnas = ['mean radius', 'mean concavity']

# Calcular IQR
Q1 = df[columnas].quantile(0.25)
Q3 = df[columnas].quantile(0.75)
IQR = Q3 - Q1

# Límites
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar el DataFrame completo
df_limpio = df[
    ~(
        (df[columnas] < limite_inferior) |
        (df[columnas] > limite_superior)
    ).any(axis=1)
].copy()

print("Filas originales:", df.shape[0])
print("Filas después del IQR:", df_limpio.shape[0])
print("Outliers eliminados:", df.shape[0] - df_limpio.shape[0])
```

# 🚀 Aplicación de agrupamiento (Clustering):

Aplicar un algoritmo de agrupamiento que permita identificar grupos o segmentos dentro del conjunto de datos. Probar con diferentes cantidades de grupos para explorar cuál podría ser una buena elección. Evaluar visualmente si los grupos
tienen sentido con base en su separación o forma.

Para esta sección vamos a usar el KMeans como el algoritmo de clustering que nos permitirá encontrar patrones en las variables seleccionadas.

##  📈 Generación de Clusters

```python
# Primero realizaremos el escalado de los datos
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Creamos un variable con las columnas seleccionadas
X = df_limpio[['mean radius', 'mean concavity']]

# Entrenamos el modelo
X_scaled = scaler.fit_transform(X)
```

```python
# Importamos la libreria del algoritmo para el entrenamiento
from sklearn.cluster import KMeans

# Definimos el número de clusters
numero_clusters = 10

# Entrenamos el modelo
est = KMeans(numero_clusters, n_init=10)
est.fit(X_scaled)

# Realizamos una predicción
y_kmeans = est.predict(X_scaled)

# Mediante el Silhouette Score para validar si el entremiento realizado tiene sentido
# Importamos la libreria
from sklearn.metrics import silhouette_score

# Ejecutamos la validación del silhouette score
score = silhouette_score(X_scaled, y_kmeans)

# Mostramos el resultado
print("Silhouette Score: ", score)
```

### 🧠 Interpretación

El silhouette Score es una métrica de evaluación de clustering para saber que tan bien está asignado cada punto a su cluster que valida, la distancia de un punto esta tan cerca de su propio cluster y la distancia de un punto esta tan cerca de otro cluster.

Un Silhouette Score de ~0.36 indica una estructura de agrupamiento (clustering) moderadamente mala e incoherente.

- El coeficiente de silueta mide qué tan similar es un punto a su propio clúster en comparación con otros clústeres, en un rango de -1 a 1, es decir, la puntuación obtenida: Significa que nuestros clústeres no están bien definidos, son internamente dispersos y se encuentran separados entre sí. Los datos están superpuestos de forma caótica.


##  📈 Gráficamos los Clusters Generados

```python
# Crear la figura
plt.figure(figsize=(10,6))

# Gráfico representativo
scatter = plt.scatter(X_scaled[: , 0], X_scaled[: , 1],
                      c= y_kmeans, # Coloreamos los puntos en base a los grupos del conjunto y_kmeans
                      cmap='tab10', # Colores para los grupos de datos
                      s = 50,
                      alpha=0.8,
                      edgecolors="black");

# Gráficamos el centroide
plt.scatter(
    est.cluster_centers_[:,0], # Coordenada X del centroide
    est.cluster_centers_[:,1], # Coordenada Y del centroide
    marker="X",
    s=300,
    c="red",
    label="Centroides"
)

# Personalización
plt.title("Distribución de Clusters Generados", fontsize=16)
plt.xlabel("Mean Radius", fontsize=14)
plt.ylabel("Mean Concavity", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)

plt.colorbar(scatter, label="Cluster");

# Mostrar el gráfico
plt.show()
```

### 🧠 Interpretación

En el gráfico nos confirma lo mencionado con la métrica de Silhouette Score que nos dice que esta muy disperso todo y no da una coherencia en los datos.

## 🔍 Análisis de Dataset por medio del método del codo

Este método nos permite realizar una visualización de los clusters en una comparación entre el número de clusters y la inercia.

```python
# Crearemos una lista para obtener la inercia de cada modelo entrenado con los datos escalados
inercias = []

# Realizamos el entrenamiento desde 1 a 11 y obtenemos su inercia
for k in range(1,11):
    modelo = KMeans(n_clusters=k, random_state=42)
    modelo.fit(X_scaled)
    inercias.append(modelo.inertia_)

# Realizamos un gráfico para visualizar el resultado
plt.plot(range(1,11), inercias, marker='o')
plt.xlabel("Número de clusters")
plt.ylabel("Inercia")
plt.title("Método del Codo")
plt.show()
```

### 🧠 Interpretación de Gráfico

Según el gráfico y el método de del codo nos dice que el número optimo de clusters es 2 por el gráfico donde se forma el codo.

```python
# -----------------------------
# Método del Codo
# -----------------------------
inercias = []

for k in range(1, 11):
    modelo = KMeans(n_clusters=k, random_state=42)
    modelo.fit(X_scaled)
    inercias.append(modelo.inertia_)

# -----------------------------
# Clustering Jerárquico
# -----------------------------
Z = linkage(X_scaled, method='ward')

# -----------------------------
# Figura con ambos gráficos
# -----------------------------
fig, ax = plt.subplots(1, 2, figsize=(16,6))

# Método del Codo
ax[0].plot(range(1,11), inercias, marker='o')
ax[0].set_title("Método del Codo")
ax[0].set_xlabel("Número de Clusters")
ax[0].set_ylabel("Inercia")
ax[0].grid(True)

# Dendrograma
dendrogram(Z, ax=ax[1])
ax[1].set_title("Dendrograma")
ax[1].set_xlabel("Observaciones")
ax[1].set_ylabel("Distancia")

plt.tight_layout()
plt.show()
```

### 🧠 Interpretación

Tanto el algoritmo K-Means como el agrupamiento jerárquico identificaron dos grupos principales de tumores. Aunque ambos métodos emplean estrategias diferentes para formar los clusters, los resultados fueron similares, indicando que las variables mean radius y mean concavity contienen una estructura de agrupamiento consistente. Las pequeñas diferencias observadas se deben a que K-Means optimiza centroides, mientras que el método jerárquico construye los grupos a partir de las distancias entre observaciones.

Al comparar ambos métodos, se observa una convergencia clara que valida y refuerza la elección de la estructura del modelo. El método del codo muestra su inflexión más drástica y significativa al pasar de 1 a 2 clústeres, donde la inercia experimenta una caída masiva (de aproximadamente 1100 a cerca de 420), a partir de la cual la reducción del error se vuelve mucho más gradual e incremental. Este comportamiento se alinea de manera óptima con el dendrograma, cuya mayor brecha vertical sin fusiones se da precisamente al cortar el árbol a una distancia de 20 para formar 2 grupos. Sumado a la métrica de Silhouette Score de aproximadamente 0.54 que obtuviste previamente, toda la evidencia analítica confirma que **$k = 2$ es el número de clústeres más sólido, equilibrado y matemáticamente justificado**, capturando la separación natural de los datos sin caer en una sobresegmentación innecesaria.
